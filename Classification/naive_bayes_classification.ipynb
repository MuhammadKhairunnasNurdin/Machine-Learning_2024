# **Naive-Bayes Theory**

Naive-Bayes is a classification algorithm based on Bayes' theorem. Bayes' theorem is a mathematical formula that calculates the probability of a hypothesis given the prior knowledge. The formula is as follows:
\begin{equation} P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} \end{equation}
Where:
- P(A|B) is the probability of event A given event B
- P(B|A) is the probability of event B given event A
- P(A) is the probability of event A
- P(B) is the probability of event B
- P(A|B) is the posterior probability
- P(B|A) is the likelihood
- P(A) is the prior probability
- P(B) is the evidence
- The denominator P(B) is the normalization constant
- The denominator P(B) is the probability of event B

For example, we will plan to go on a picnic.  Current conditions are cloudy.  We also get the fact that,

 o 50% of rainy days begin with cloudy weather

 o Cloudy conditions often occur in the morning with a 40% chance

 o General conditions currently tend to be dry and rain is rare.  Rain only occurred 3 days in 30 days.

 Then we want to know the chance of rain this afternoon before we decide to go on a picnic.

 Based on these facts, we can formulate it in Bayes form, namely,

 o Determine event A and event B. In this case, we assume that event A is rain, while event B is cloudy conditions.  So we get P(Rain)=P(A) and P(Cloudy)=P(B).

 o 50% of rainy days start with cloudy weather  P(Cloudy│Rainy)=P(B│A)=50%=0.5.

 o Cloudy conditions often occur in the morning with a 40% chance  P(B)=40%=0.4

 o Dry conditions and rare rain can be modeled as P(A) with a probability of 3/30=10%=0.1 P(A)=0.1

 o So to calculate the probability of rain during the day, we can model it as P(A|B)

 The obtained value of P(A|B) is: P(A|B)=P(B|A) x P(A) / P(B) = 0.5 x 0.1 / 0.4 = 0.125


 Thus, the chance of rain during the day is 12.5%. 

So what relation with naive bayes model?, In the naïve Bayes model we can calculate the probability of a label in a class in a classification case based on the probability of each feature for that label.  So, to calculate the probability of label y based on features x_n in naïve Bayes, this is done using the formula:
\begin{equation} P(y|x_1, x_2, ..., x_n) = \frac{P(y) \cdot P(x_1|y) \cdot P(x_2|y) \cdot ... \cdot P(x_n|y)}{P(x_1) \cdot P(x_2) \cdot ... \cdot P(x_n)} \end{equation}
Where:
- P(y|x_1, x_2, ..., x_n) is the probability of label y given features x_1, x_2, ..., x_n
- P(y) is the probability of label y
- P(x_1|y) is the probability of feature x_1 given label y
- P(x_2|y) is the probability of feature x_2 given label y
- P(x_n|y) is the probability of feature x_n given label y

After getting the probability of label \( y \) based on \( x_n \) features, the final process of the Naive Bayes model is to compare the probability for each label. The label with the greatest probability is the prediction result from the Naive Bayes model. 

Here’s an example in the case of predicting whether an email is spam or not spam:

- **Scenario**: Classifying emails as **Spam** or **Not Spam** based on content.

- **Training Data**:
    - "Win a free iPhone" → Spam
    - "Limited time offer" → Spam
    - "Meeting at 10 AM" → Not Spam
    - "Your invoice is ready" → Not Spam
    - "Congratulations, you've won!" → Spam

- **Features**: 
    - Words: "win", "free", "iphone", "limited", "time", "offer", "meeting", "invoice", "congratulations"

- **Step-by-Step Process**:
    1. **Prior Probabilities**:
        - Spam probability: 
        $$ P(\text{Spam}) = \frac{3}{5} = 0.6 $$
        - Not Spam probability: 
        $$ P(\text{Not Spam}) = \frac{2}{5} = 0.4 $$

    2. **Probability of Each Word**:
        Note that stop words are not considered. Stop words are commonly used words in the English language, such as "the", "a", "an", "in", etc. The process for calculating the probability of the word "win" with respect to the sentence "Win a free iPhone" is as follows: since "win" appears 2 times in 3 spam sentences, we have:
        - "win": 
        $$ P(\text{"win"} | \text{Spam}) = \frac{2}{3} \quad (0.67), \quad P(\text{"win"} | \text{Not Spam}) = 0 $$
        - "free": 
        $$ P(\text{"free"} | \text{Spam}) = \frac{1}{3} \quad (0.33), \quad P(\text{"free"} | \text{Not Spam}) = 0 $$
        - "iphone": 
        $$ P(\text{"iphone"} | \text{Spam}) = \frac{1}{3} \quad (0.33), \quad P(\text{"iphone"} | \text{Not Spam}) = 0 $$
        - "limited": 
        $$ P(\text{"limited"} | \text{Spam}) = \frac{1}{3} \quad (0.33), \quad P(\text{"limited"} | \text{Not Spam}) = 0 $$
        - "time": 
        $$ P(\text{"time"} | \text{Spam}) = \frac{1}{3} \quad (0.33), \quad P(\text{"time"} | \text{Not Spam}) = 0 $$
        - "offer": 
        $$ P(\text{"offer"} | \text{Spam}) = \frac{1}{3} \quad (0.33), \quad P(\text{"offer"} | \text{Not Spam}) = 0 $$
        - "meeting": 
        $$ P(\text{"meeting"} | \text{Spam}) = 0, \quad P(\text{"meeting"} | \text{Not Spam}) = \frac{1}{2} \quad (0.5) $$
        - "invoice": 
        $$ P(\text{"invoice"} | \text{Spam}) = 0, \quad P(\text{"invoice"} | \text{Not Spam}) = \frac{1}{2} \quad (0.5) $$
        - "congratulations": 
        $$ P(\text{"congratulations"} | \text{Spam}) = \frac{1}{3} \quad (0.33), \quad P(\text{"congratulations"} | \text{Not Spam}) = 0 $$

    3. **New Email to Classify**: "Win a free iPhone"
    
    4. **Posterior Probabilities**:
        - For Spam: 
        $$ P(\text{Spam} | \text{"Win", "free", "iPhone"}) \propto P(\text{Spam}) \times P(\text{"win"} | \text{Spam}) \times P(\text{"free"} | \text{Spam}) \times P(\text{"iphone"} | \text{Spam}) $$
        $$ \approx 0.6 \times 0.67 \times 0.33 \times 0.33 \approx 0.0402 $$
        - For Not Spam: 
        $$ P(\text{Not Spam} | \text{"Win", "free", "iPhone"}) = 0 $$

    5. **Final Prediction**:
        - Probabilities:
            - Spam: 
            $$ P(\text{Spam}) \approx 0.0402 $$
            - Not Spam: 
            $$ P(\text{Not Spam}) = 0 $$
    
    6. **Result**: The email "Win a free iPhone" is classified as **Spam**.

- **Summary**: The Naive Bayes model predicts the label based on the highest probability derived from the features (words) present in the email, here those example compare to actual formula of naive bayes model:
In the Naive Bayes model, the full Bayes' theorem is:
$$ 
P(y|x_1, x_2, \ldots, x_n) = \frac{P(y) \cdot P(x_1|y) \cdot P(x_2|y) \cdots P(x_n|y)}{P(x_1) \cdot P(x_2) \cdots P(x_n)} 
$$
**Why We Omit $$ ( P(x_1) \cdot P(x_2) \cdots P(x_n) ) $$ here explanation**
    1. **Normalization Purpose**:
        - The term $$ ( P(x_1) \cdot P(x_2) \cdots P(x_n) ) $$ serves to normalize the posterior probability. It ensures that the probabilities sum to 1 across all possible classes.
        - However, when comparing probabilities between different classes (e.g., Spam vs. Not Spam), this normalization factor is constant for all classes, so it does not affect the relative comparison of which class has the highest probability.
    
    2. **Relative Comparison**:
        - In practice, we often care about which class \( y \) has the highest posterior probability. Since the normalization factor is the same for all classes, it can be omitted when calculating the relative probabilities:
        $$
        P(y|x_1, x_2, \ldots, x_n) \propto P(y) \cdot P(x_1|y) \cdot P(x_2|y) \cdots P(x_n|y)
        $$
        - This proportionality allows us to simplify calculations without affecting the outcome of which class has the highest probability.

**Example Application**

In your example for classifying the email "Win a free iPhone":

- **Posterior for Spam**:
$$
P(\text{Spam} | \text{"Win", "free", "iPhone"}) \propto P(\text{Spam}) \cdot P(\text{"win"} | \text{Spam}) \cdot P(\text{"free"} | \text{Spam}) \cdot P(\text{"iphone"} | \text{Spam})
$$

- **Posterior for Not Spam**:
$$
P(\text{Not Spam} | \text{"Win", "free", "iPhone"}) \propto P(\text{Not Spam}) \cdot P(\text{"win"} | \text{Not Spam}) \cdot P(\text{"free"} | \text{Not Spam}) \cdot P(\text{"iphone"} | \text{Not Spam})
$$

The naïve Bayes model is called "naïve" because it assumes that each feature is independent of each other.  This assumption simplifies the calculation process and makes the model easier to implement. Also with this assumption, the model can still provide good results, especially for text classification and text filtering problems, but if one feature added, you had to calculate the probability of the label based on the new feature.  So, the model can be updated continuously with new data, but you had calculate again the probability of the label based on the new feature.  This is the disadvantage of the naive bayes model, but it is still widely used in text classification problems and text filtering.

# **Create Dummy(syntetic) dataset**

```python
# Importing necessary libraries
# make_classification is used to create synthetic dataset
import numpy as np
from sklearn.datasets import make_classification

# Create a synthetic dataset
# n_samples: number of samples
# n_features: number of features
# n_classes: number of classes
# n_informative: number of informative features 
# , informative features are the features that actually influence the target variable
# n_redundant: number of redundant features
# ,redundant features are the features that are derived from the informative features and not had correlation with class
# this usually done to increase the complexity of the dataset
# n_repeated: number of duplicated features
# , duplicated features are the features that are copied from the informative features
# shuffle: shuffle the samples
X, y = make_classification(n_samples=30, n_features=2, n_classes=2, n_informative=2, n_redundant=0, n_repeated=0, shuffle=False)

# by default make_classification return float64 data type
# convert to discrete value 
# absolute function is used to convert negative values to positive
X = np.absolute(X)
# round function is used to round the values to 2 decimal places
# multiply by 100 to convert the values to percentage
# so no more decimal points
X = np.round(X, 2) * 100
# convert the float64 to int
X = X.astype(int)
print(X)
print(y)
```

# **Create DataFrame**
To make the data easier to read, we will create a DataFrame using the Pandas library based on dummy data that was created previously. 

```python
# Importing necessary libraries
import pandas as pd

# reshape label y to 2D array
# this is done to concatenate the X and y
y = y.reshape(len(y), 1)
# concatenate X and y in column wise to create a single dataset
data = np.concatenate((X, y), axis=1)
# naming the columns
columns = ['Feature 1', 'Feature 2', 'Label']
# create a DataFrame
df = pd.DataFrame(data, columns=columns)
# display the first 5 rows of the DataFrame
df.head()
```

# **Convert Label to categorical**

```python
# before define label name, you can check the unique value of the label
# df['Label'].unique()

# define label name
labels = {
    0: 'Class B',
    1: 'Class A'
}

# copy the original DataFrame to labeled DataFrame
df_label = df.copy()
# replace the label values with the label name above
# using map() function in pandas
df_label['Label'] = df_label['Label'].map(labels)

# display the first 5 rows of the labeled DataFrame
df_label.head()
```

# **Visualize the dataset**

```python
# Importing necessary libraries
import matplotlib.pyplot as plt

# define the colors for each class
colors = {
    'class_a': 'MediumVioletRed',
    'class_b': 'Navy'
}
# group the data based on the label
grouped = df_label.groupby('Label')
class_a = grouped.get_group('Class A')
class_b = grouped.get_group('Class B')

# plot the data
plt.scatter(x=class_a['Feature 1'], y=class_a['Feature 2'], color=colors['class_a'], )
plt.scatter(x=class_b['Feature 1'], y=class_b['Feature 2'], color=colors['class_b'], label='Feature 2')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend(['Class A', 'Class B'], loc='upper left')
plt.gca().axes.xaxis.set_ticklabels([])
plt.gca().axes.yaxis.set_ticklabels([])
plt.show()
```

# **Multinomial Naive-Bayes Classification**
Theory:
- **Multinomial Naive-Bayes** is a variant of the Naive-Bayes algorithm that is used for classification tasks.
- It is suitable for classification tasks with discrete features (e.g., word counts for text classification).
- The Multinomial Naive-Bayes model is based on the assumption that the features are generated from a simple multinomial distribution
- multinomial distribution is a generalization of the binomial distribution to more than two categories, where each feature represents the frequency of a term in a document, means the frequency of a word in a text, like in our example, the frequency of the word "win" in a spam email or the frequency of the word "meeting" in a non-spam email.
- The model calculates the probability of each label based on the probability of each feature for that label.
- The label with the highest probability is the predicted label for the input data.
- The Multinomial Naive-Bayes model is widely used in text classification tasks, such as spam detection and sentiment analysis.

```python
# Importing necessary libraries
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# initiate object Multinomial Naive-Bayes
mnb = MultinomialNB()

# convert the label to 1D array
y = y.ravel()

# Use the synthetic dataset that was created previously
# X: features dataset (Feature 1 and Feature 2)
# y: label dataset (Class A and Class B)
# split the data into training and testing set
# test_size: the proportion of the dataset to include in the test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=30)

# train or fit the model
mnb.fit(X_train, y_train)

# predict the label of the training set
y_train_pred = mnb.predict(X_train)
# evaluate the accuracy of the model
train_accuracy = accuracy_score(y_train, y_train_pred)

# predict the label of the testing set
y_test_pred = mnb.predict(X_test)
# evaluate the accuracy of the model
test_accuracy = accuracy_score(y_test, y_test_pred)

# display the accuracy of the model
print(f'Training Accuracy: {train_accuracy}')
print(f'Testing Accuracy: {test_accuracy}')
```

# **Gaussian Naive-Bayes Classfification**
Theory:
- **Gaussian Naive-Bayes** is a variant of the Naive-Bayes algorithm that is used for classification tasks.
- It is suitable for classification tasks with continuous features (e.g., real-valued features).
- The Gaussian Naive-Bayes model is based on the assumption that the features are generated from a simple Gaussian distribution.
- The model calculates the probability of each label based on the probability of each feature for that label.
- The label with the highest probability is the predicted label for the input data.
- The model can be updated continuously with new data, making it suitable for real-time applications.
- The Gaussian Naive-Bayes model is robust to noise and can handle missing values in the dataset.
- The model is suitable for large datasets and can provide good results even with limited training data.
- Disadvantages of the Gaussian Naive-Bayes model include the assumption of feature independence, which may not hold in some cases, and the model's sensitivity to the distribution of the data.
- The model may not perform well with highly correlated features or features with complex relationships.
- The Gaussian Naive-Bayes model is widely used in classification tasks with continuous features, such as medical diagnosis and image recognition.

```python
# Importing necessary libraries
from sklearn.naive_bayes import GaussianNB

# initiate object Gaussian Naive-Bayes
gnb = GaussianNB()

# Use the synthetic dataset that was created previously
# and Use the same training and testing set that was created previously
# train or fit the model
gnb.fit(X_train, y_train)

# predict the label of the training set
y_train_pred_gnb = gnb.predict(X_train)
# evaluate the accuracy of the model
train_accuracy_gnb = accuracy_score(y_train, y_train_pred_gnb)

# predict the label of the testing set
y_test_pred_gnb = gnb.predict(X_test)
# evaluate the accuracy of the model
test_accuracy_gnb = accuracy_score(y_test, y_test_pred_gnb)

# display the accuracy of the model
print(f'Training Accuracy: {train_accuracy_gnb}')
print(f'Testing Accuracy: {test_accuracy_gnb}')
```

# **Conclusion**
**When to Use Multinomial Naive Bayes (MNB)**
------------------------------------------------

1. **Discrete Features**:
    - MNB is ideal for datasets where features are represented as counts or frequencies. Common examples include:
        - **Text Classification**: In text processing, each document can be represented by a vector of word counts. MNB evaluates the likelihood of each word appearing in a document based on its frequency, making it suitable for tasks like:
            - **Spam Detection**: Classifying emails as spam or not spam based on the frequency of certain keywords (e.g., "free," "win").
            - **Sentiment Analysis**: Determining the sentiment of reviews (positive, negative, neutral) by analyzing the occurrence of sentiment-laden words.
        - **Categorical Features**: MNB can also be applied to datasets where features represent counts of categorical variables, such as the number of different products purchased by a customer.

2. **Natural Language Processing (NLP)**:
    - MNB is particularly effective for text classification tasks, such as:
        - **Topic Categorization**: Automatically categorizing news articles into topics (e.g., sports, politics) based on the frequency of relevant terms.
        - **Intent Classification**: In chatbots, MNB can help classify user intents by analyzing the frequency of words in user queries.

3. **Small Sample Sizes**:
    - MNB can perform well even with relatively small amounts of data, as it uses frequency counts rather than relying on the distribution of continuous variables. This makes it useful in scenarios where collecting large datasets is challenging.

**When to Use Gaussian Naive Bayes (GNB)**
---------------------------------------------

1. **Continuous Features**:
    - GNB is suitable for datasets where features are continuous and can be assumed to follow a Gaussian (normal) distribution. Common examples include:
        - **Measurement Data**: Features derived from physical measurements such as height, weight, or temperature. For instance, in healthcare, GNB can classify patients based on continuous variables like blood pressure and cholesterol levels.

2. **Mixed Data Types**:
    - If your dataset contains both continuous and categorical features, GNB can be applied to the continuous features while you can use MNB for the categorical ones. This is often done in hybrid models. For example:
        - **Customer Segmentation**: In marketing, you might have continuous features (e.g., age, income) and categorical features (e.g., preferred product category). You can use GNB for the continuous features and MNB for the categorical features to segment customers effectively.

3. **Assumption of Normality**:
    - GNB assumes that the continuous features are normally distributed within each class. If this assumption holds true for your data, GNB can yield good results. For example:
        - **Credit Scoring**: In finance, GNB can classify loan applicants based on features like income, credit score, and age, assuming these features are normally distributed across accepted and rejected applicants.

**Summary of Use Cases**
---------------------------

| **Feature Type**        | **Model**             | **Use Cases**                                     |
|-------------------------|-----------------------|---------------------------------------------------|
| Discrete counts         | Multinomial Naive Bayes | Text classification (spam, sentiment analysis)   |
| Categorical features     | Multinomial Naive Bayes | Count data and frequency-based tasks              |
| Continuous data         | Gaussian Naive Bayes  | Measurement data (height, weight, etc.)          |
| Mixed data              | Combination of both   | Use MNB for categorical and GNB for continuous features |

**Summary**
---------------

In summary, choose **Multinomial Naive Bayes** for discrete, count-based data (especially in text classification) and **Gaussian Naive Bayes** for continuous data that follows a normal distribution. Understanding the nature of your features will guide you in selecting the appropriate model for your classification tasks. By assessing the characteristics of your data, you can make an informed decision on which Naive Bayes variant to utilize for optimal performance.
