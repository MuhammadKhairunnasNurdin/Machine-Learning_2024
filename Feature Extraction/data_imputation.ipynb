# *Load Data* 

```python
import pandas as pd

# path for dataset
data = 'Data/Titanic-Dataset.csv'
# load dataset
# df for acronym of DataFrame
df = pd.read_csv(data)
# show the first 5 rows of dataset
df.head()
```

# *Checking Data*

```python
# info() function in pandas can show 
# amount of data and data-type per-column
df.info()

# isnull().sum() in pandas can show
# what's data that missing per-column
df.isnull().sum()
```

# *Imputation*

```python
# Imputation is a process to fill missing data
# because inplace=True cannot do in this case pandas 3.0
# so we need to assign the value to the column itself
# we use mean, mode, or median to fill the missing data

# Because Age is nominal data
# we use mean strategic
df['Age'] = df['Age'].fillna(value=df['Age'].mean())

# Because we assume that ship passengers 
# don't had cabin number are in ship deck
# so we substitute that with "DECK" value
df['Cabin']= df['Cabin'].fillna(value="DECK")

# Because Embarked is embark location
# so there are two column that missing
# we can use mode to substitute that
df['Embarked']= df['Embarked'].fillna(value=df['Embarked'].mode().iloc[0])

# show Embarked data after imputation
df['Embarked']
```
